```python
import seaborn as sn
```

```python
import matplotlib.pyplot as pl
```

```python
g=data.groupby('country')['total_bill'].mean().reset_index().sort_values(by='total_bill',ascending=False).head(10)
```

```python
g
```

```python
pl.figure(figsize=(15,4))
a=sn.barplot(data=g,x='country',y='total_bill',hue='country')
for i in a.containers:
  a.bar_label(i)
pl.title('top 10 country WTR of total bill')
pl.show()
```

```python
g=data.groupby('country')['invoice_frequency'].mean().reset_index().sort_values(by='invoice_frequency',ascending=False).head(10)
```

```python
pl.figure(figsize=(15,4))
a=sn.barplot(data=g,x='country',y='invoice_frequency',hue='country')
for i in a.containers:
  a.bar_label(i)
pl.title('top 10 country WTR of invoice_frequency')
pl.show()
```
